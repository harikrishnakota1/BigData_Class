Peform the following array operations using numpy   
( show both operational use of numpy and functional (example addition operation => + vs addition function => numbpy.add() )

### 3.3.1 addition a+b 

```python
# Operational use
import numpy as np
a=np.array(([1,2,3,4]), dtype=float)
b=np.array(([5,6,7,8]),dtype=float)
a+b
```

```python
# Functional use

a=array([1,2,3,4])
b=array([5,6,7,8])
c = np.add((a),(b), dtype=float )
c
```

### 3.3.2 subtraction a-b

```python
# Operational use

a=np.array(([1,2,3,4]), dtype=float)
b=np.array(([5,6,7,8]),dtype=float)
a-b
```

```python
# Functional use

a=array([1,2,3,4])
b=array([5,6,7,8])
c = np.subtract((a),(b), dtype=float )
c
```
